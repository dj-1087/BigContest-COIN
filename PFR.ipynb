# 개인

```python
def load_csv(PATH, season):
    doc_ip = None
    for i in season:
        doc_i = pd.read_csv(PATH+str(i)+".csv")
        doc_ip = pd.concat([doc_ip,doc_i], axis=0)
    doc_ip = doc_ip.set_index(pd.Index(range(0,doc_ip.shape[0])))
    return doc_ip
df = load_csv(PATH_IP, season)
df
```

# PFR: (삼진+볼넷)/이닝

```python
def convert_PFR(df):
    needs = df.copy()
    
    for i, row in needs.iterrows():
        needs.at[i,"G_ID"] = needs.at[i,"G_ID"][:4]
    needs = needs.groupby(by=["G_ID","P_ID"]).sum()
    needs = needs[["KK","BB","INN2"]]
    needs["INN2"] = needs["INN2"]/3
    pfr = (needs["KK"]+needs["BB"])/needs["INN2"]
    return pfr
pfr = convert_PFR(df)
pfr
```

```python
def load_csv(PATH, season):
    doc_tp = None
    for i in season:
        doc_i = pd.read_csv(PATH+str(i)+".csv")
        doc_tp = pd.concat([doc_tp,doc_i], axis=0)
    doc_tp = doc_tp.set_index(pd.Index(range(0,doc_tp.shape[0])))
    return doc_tp
team_df = load_csv(PATH_TP, season)
team_df
```

```python
def convert_team_PFR(df):
    needs = df.copy()
    
    for i, row in needs.iterrows():
        needs.at[i,"G_ID"] = needs.at[i,"G_ID"][:4]
    needs = needs.groupby(by=["G_ID","T_ID"]).sum()
    needs = needs[["KK","BB","INN2"]]
    needs["INN2"] = needs["INN2"]/3
    pfr = (needs["KK"]+needs["BB"])/needs["INN2"]
    return pfr
team_pfr = convert_team_PFR(team_df)
team_pfr
```

```python
def pfr(row):
    previous = team_pitcher[(team_pitcher['시즌'] == row['시즌']) & (team_pitcher['팀코드'] == row['팀코드']) & (team_pitcher['일자'] < row['일자'])]
    if len(previous) == 0:
        row['PFR'] = 0
        return row
    previous_sum = previous.sum()
    try:
        pg = (previous_sum["삼진"]+previous_sum["4구"])/(previous_sum["이닝*3"]/3)
    except ZeroDivisionError:
        row['PFR'] = 0
    row["PFR"] = pg
    return row
def perPFR(row):
    previous = pitcher[pitcher['선수코드'] == row['선수코드']].copy()
    previous = previous[(previous['시즌'] == row['시즌']) &(previous['일자'] < row['일자'])]
    previous_sum = previous.sum()
    if len(previous) == 0:
        row['PFR'] = 0
        return row
    try:
        pg = (previous_sum["삼진"]+previous_sum["4구"])/(previous_sum["이닝*3"]/3)
    except ZeroDivisionError:
        row['PFR'] = 0
        return row
    row['PFR'] = pg
    return row
```

```python
team_pitcher = team_pitcher.apply(pfr,axis=1)
team_pitcher["PFR"].tail()
```

```python
pitcher = pitcher.apply(perPFR,axis=1)
pitcher["PFR"]
```

```python
team_pitcher.to_csv("팀투수_세이버 매트릭스_합본.csv",encoding='utf-8-sig')
pitcher.to_csv("개인투수_세이버 매트릭스_합본.csv",encoding='utf-8-sig')
```
