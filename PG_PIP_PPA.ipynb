```python
import pandas as pd
#ip = individual pitcher
PATH_IP = "C:/Users/home/DA(inflearn)/individual_pitcher/2020빅콘테스트 문제데이터(데이터분석분야-퓨처스리그)/2020빅콘테스트 데이터분석분야-퓨쳐스리그_스포츠투아이_제공데이터(.CSV)_시즌별, 시트별 구분/2020빅콘테스트_스포츠투아이_제공데이터_개인투수_"
#tp = team pitcher
PATH_TP = "C:/Users/home/DA(inflearn)/individual_pitcher/2020빅콘테스트 문제데이터(데이터분석분야-퓨처스리그)/2020빅콘테스트 데이터분석분야-퓨쳐스리그_스포츠투아이_제공데이터(.CSV)_시즌별, 시트별 구분/2020빅콘테스트_스포츠투아이_제공데이터_팀투수_"
season = [2016, 2017, 2018, 2019, 2020]
```

# 개인

```python
def load_csv(PATH, season):
    doc_ip = None
    for i in season:
        doc_i = pd.read_csv(PATH+str(i)+".csv")
        doc_ip = pd.concat([doc_ip,doc_i], axis=0)
    doc_ip = doc_ip.set_index(pd.Index(range(0,doc_ip.shape[0])))
    return doc_ip
df = load_csv(PATH_IP, season)
df
```

# P/G: 경기당 투구수 

```python
def convert_PG(df):
    needs = df.copy()
    
    for i, row in needs.iterrows():
        needs.at[i,"season"] = needs.at[i,"G_ID"][:4]
        needs.at[i,"count"] = 1
    needs = needs.groupby(by=["season","P_ID"]).sum()
    needs = needs[["BF", "count"]]
    pg = needs["BF"]/needs["count"]
    return pg
pg = convert_PG(df)
pg
```

# P/IP: 이닝당 투구수

```python
def convert_PIP(df):
    needs = df.copy()
    
    for i, row in needs.iterrows():
        needs.at[i,"G_ID"] = needs.at[i,"G_ID"][:4]
    needs = needs.groupby(by=["G_ID","P_ID"]).sum()
    needs = needs[["INN2","BF"]]
    needs["INN2"] = needs["INN2"]/3
    pip = needs["BF"]/needs["INN2"]
    return pip
p_ip = convert_PIP(df)
p_ip
```

# P/PA: 타석당 투구수

```python
def convert_PPA(df):
    needs = df.copy()
    
    for i, row in needs.iterrows():
        needs.at[i,"G_ID"] = needs.at[i,"G_ID"][:4]
    needs = needs.groupby(by=["G_ID","P_ID"]).sum()
    needs = needs[["BF","PA"]]
    ppa = needs["BF"]/needs["PA"]
    return ppa
p_pa = convert_PPA(df)
p_pa
```

```python
def pg(row):
    previous = team_pitcher[(team_pitcher['시즌'] == row['시즌']) & (team_pitcher['팀코드'] == row['팀코드']) & (team_pitcher['일자'] < row['일자'])]
    if len(previous) == 0:
        row['P/G'] = 0
        return row
    previous_sum = previous.sum()
    try:
        pg = previous_sum["투구수"]/previous.count()[0]
    except ZeroDivisionError:
        row['P/G'] = 0
    row["P/G"] = pg
    return row
def perPG(row):
    previous = pitcher[pitcher['선수코드'] == row['선수코드']].copy()
    previous = previous[(previous['시즌'] == row['시즌']) &(previous['일자'] < row['일자'])]
    previous_sum = previous.sum()
    if len(previous) == 0:
        row['P/G'] = 0
        return row
    try:
        pg = previous_sum["투구수"]/previous.count()[0]
    except ZeroDivisionError:
        row['P/G'] = 0
        return row
    row['P/G'] = pg
    return row
```
