```python
import pandas as pd
TEAM_PATH = "C:/Users/home/DA(inflearn)/individual_pitcher/팀투수_세이버 매트릭스_합본_8_25.csv"
PER_PATH = "C:/Users/home/DA(inflearn)/individual_pitcher/개인투수_세이버 매트릭스_합본_8_25.csv"
```

```python
team_pitcher = pd.read_csv(TEAM_PATH,encoding="utf-8-sig")
team_pitcher
```

```python
per_pitcher = pd.read_csv(PER_PATH,encoding="utf-8-sig")
per_pitcher
```

```python
# 안타/타수
def teamOAVG(row):
    previous = team_pitcher[(team_pitcher['시즌'] == row['시즌']) & (team_pitcher['팀코드'] == row['팀코드']) & (team_pitcher['일자'] < row['일자'])]
    if len(previous) == 0:
        row['OAVG'] = 0
        return row
    previous_sum = previous.sum()
    try:
        oavg = previous_sum["안타"]/previous_sum["타수"]
    except ZeroDivisionError:
        oavg = 0
    row['OAVG']= oavg
    return row
def perOAVG(row):
    previous = per_pitcher[per_pitcher['선수코드'] == row['선수코드']].copy()
    previous = previous[(previous['시즌'] == row['시즌']) &(previous['일자'] < row['일자'])]
    previous_sum = previous.sum()
    if len(previous) == 0:
        row['OAVG'] = 0
        return row
    try:
        oavg = previous_sum["안타"]/previous_sum["타수"]
    except ZeroDivisionError:
        row['OAVG'] = 0
        return row
    row['OAVG'] = oavg
    return row

#출루율 = 안타+(4구+사구)/타수+(4구+사구)+희비
def teamOOBP(row):
    previous = team_pitcher[(team_pitcher['시즌'] == row['시즌']) & (team_pitcher['팀코드'] == row['팀코드']) & (team_pitcher['일자'] < row['일자'])]
    if len(previous) == 0:
        row['OOBP'] = 0
        return row
    previous_sum = previous.sum()
    try:
        oobp = (previous_sum["안타"]+previous_sum["4구"]+previous_sum["사구"])/(previous_sum["타수"]+previous_sum["4구"]+previous_sum["사구"]+previous_sum["희비"])
    except ZeroDivisionError:
        oobp = 0
    row['OOBP']= oobp
    return row
def perOOBP(row):
    previous = per_pitcher[per_pitcher['선수코드'] == row['선수코드']].copy()
    previous = previous[(previous['시즌'] == row['시즌']) &(previous['일자'] < row['일자'])]
    previous_sum = previous.sum()
    if len(previous) == 0:
        row['OOBP'] = 0
        return row
    try:
        oobp = (previous_sum["안타"]+previous_sum["4구"]+previous_sum["사구"])/(previous_sum["타수"]+previous_sum["4구"]+previous_sum["사구"]+previous_sum["희비"])
    except ZeroDivisionError:
        row['OOBP'] = 0
        return row
    row['OOBP'] = oobp
    return row

#장타율 = [단타 + (2*2루타) + (3*3루타) + (4*홈런] / 타수
def teamOSLG(row):
    previous = team_pitcher[(team_pitcher['시즌'] == row['시즌']) & (team_pitcher['팀코드'] == row['팀코드']) & (team_pitcher['일자'] < row['일자'])]
    if len(previous) == 0:
        row['OSLG'] = 0
        return row
    previous_sum = previous.sum()
    try:
        oslg = (previous_sum["단타"]+2*previous_sum["2루타"]+3*previous_sum["3루타"]+4*previous_sum["홈런"])/previous_sum["타수"]
    except ZeroDivisionError:
        oslg = 0
    row['OSLG']= oslg
    return row
def perOSLG(row):
    previous = per_pitcher[per_pitcher['선수코드'] == row['선수코드']].copy()
    previous = previous[(previous['시즌'] == row['시즌']) &(previous['일자'] < row['일자'])]
    previous_sum = previous.sum()
    if len(previous) == 0:
        row['OSLG'] = 0
        return row
    try:
        oslg = (previous_sum["단타"]+2*previous_sum["2루타"]+3*previous_sum["3루타"]+4*previous_sum["홈런"])/previous_sum["타수"]
    except ZeroDivisionError:
        row['OSLG'] = 0
        return row
    row['OSLG'] = oslg
    return row
#OPS = 장타율 + 출루율
def teamOOPS(row):
    previous = team_pitcher[(team_pitcher['시즌'] == row['시즌']) & (team_pitcher['팀코드'] == row['팀코드']) & (team_pitcher['일자'] < row['일자'])]
    if len(previous) == 0:
        row['OOPS'] = 0
        return row
    previous_sum = previous.sum()
    try:
        oops = previous_sum["OOBP"]+previous_sum["OSLG"]
    except ZeroDivisionError:
        oops = 0
    row['OOPS']= oops
    return row
def perOOPS(row):
    previous = per_pitcher[per_pitcher['선수코드'] == row['선수코드']].copy()
    previous = previous[(previous['시즌'] == row['시즌']) &(previous['일자'] < row['일자'])]
    previous_sum = previous.sum()
    if len(previous) == 0:
        row['OOPS'] = 0
        return row
    try:
        oops = previous_sum["OOBP"]+previous_sum["OSLG"]
    except ZeroDivisionError:
        row['OOPS'] = 0
        return row
    row['OOPS'] = oops
    return row
```

```python
team_pitcher = team_pitcher.apply(teamOAVG,axis=1)
team_pitcher["OAVG"].iloc[2870: 2890]
```

```python
def singleHit(df):
    df["단타"] = df["안타"]-df["2루타"]-df["3루타"]-df["홈런"]
    return df
```
