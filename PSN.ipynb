# PSN = 2 x ( 홈런 x 도루 ) / ( 홈런 + 도루 )

```python
import pandas as pd
TEAM_PATH = "C:/Users/home/DA(inflearn)/individual_pitcher/팀타자_세이버 매트릭스_합본_8_25.csv"
PER_PATH = "C:/Users/home/DA(inflearn)/individual_pitcher/개인타자_세이버 매트릭스_합본_8_25.csv"
```

```python
team_hitter = pd.read_csv(TEAM_PATH,encoding="utf-8-sig")
team_hitter
```

```python
team_hitter.columns
```

```python
hitter = pd.read_csv(PER_PATH,encoding="utf-8-sig")
hitter
```

```python
hitter.columns
```

```python
# 2 x ( 홈런 x 도루 ) / ( 홈런 + 도루 )
def psn(row):
    previous = team_hitter[(team_hitter['시즌'] == row['시즌']) & (team_hitter['팀코드'] == row['팀코드']) & (team_hitter['일자'] < row['일자'])]
    if len(previous) == 0:
        row['PSN'] = 0
        return row
    previous_sum = previous.sum()
    try:
        psn = 2*(previous_sum["홈런"]*previous_sum["도루"])/(previous_sum["홈런"]+previous_sum["도루"])
    except ZeroDivisionError:
        psn = 0
    row['PSN']= psn
    return row
def perPSN(row):
    previous = hitter[hitter['선수코드'] == row['선수코드']].copy()
    previous = previous[(previous['시즌'] == row['시즌']) &(previous['일자'] < row['일자'])]
    previous_sum = previous.sum()
    if len(previous) == 0:
        row['PSN'] = 0
        return row
    try:
        psn = 2*(previous_sum["홈런"]*previous_sum["도루"])/(previous_sum["홈런"]+previous_sum["도루"])
    except ZeroDivisionError:
        row['PSN'] = 0
        return row
    row['PSN'] = psn
    return row
```

```python
team_hitter = team_hitter.apply(psn,axis=1)
team_hitter["PSN"].iloc[2870: 2890]
```

```python
hitter = hitter.apply(perPSN,axis=1)
hitter["PSN"]
```
