```
def after_may(df):
    team_list = []
    for year in range(2016, 2021):
        a = df[(df['GDAY_DS'] > '{}-05-10'.format(year)) & (df['GYEAR'] == year)]
        team_list.append(a)
    df = pd.concat(team_list)
    return df
```

```
team_hitter = after_may(team_hitter)
team_hitter = team_hitter.reset_index()
team_hitter
```

```
team_pitcher = after_may(team_pitcher)
team_pitcher = team_pitcher.reset_index()
team_pitcher
```

```
team_hitter = team_hitter.drop("RUN",axis=1)
team_pitcher = team_pitcher.drop("R", axis=1)
```

```
def binaryData(df):
    for i in range(0,len(df),2):
        for j in range(0,len(df.columns)):
            if type(df.iloc[i,j])==str:
                continue
            if df.iloc[i,j]>df.iloc[i+1,j]:
                df.iloc[i,j] = 1
            else:
                df.iloc[i,j] = 0
    df = df.drop(range(1,len(df),2))
    return df
```

```
team_hitter = binaryData(team_hitter)
team_pitcher = binaryData(team_pitcher)
```
