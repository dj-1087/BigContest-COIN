```python
import pandas as pd
TEAM_PATH = "C:/Users/home/DA(inflearn)/individual_pitcher/팀투수_세이버 매트릭스_합본.csv"
PER_PATH = "C:/Users/home/DA(inflearn)/individual_pitcher/개인투수_세이버 매트릭스_합본.csv"
```

```python
team_pitcher = pd.read_csv(TEAM_PATH)
team_pitcher[["RA/9","평균자책점"]]
```

```python
team_pitcher.columns
```

```python
pitcher = pd.read_csv(PER_PATH)
pitcher
```

```python
pitcher.columns
```

```python
def fe(row):
    previous = team_pitcher[(team_pitcher['시즌'] == row['시즌']) & (team_pitcher['팀코드'] == row['팀코드']) & (team_pitcher['일자'] < row['일자'])]
    if len(previous) == 0:
        row['F-E'] = 0
        return row
    previous_mean = previous.mean()
    try:
        fe = previous_mean["FIP"]-previous_mean["평균자책점"]
    except ZeroDivisionError:
        row['F-E'] = 0
    row["F-E"] = fe
    return row
def perFE(row):
    previous = pitcher[pitcher['선수코드'] == row['선수코드']].copy()
    previous = previous[(previous['시즌'] == row['시즌']) &(previous['일자'] < row['일자'])]
    previous_mean = previous.mean()
    if len(previous) == 0:
        row['F-E'] = 0
        return row
    try:
        fe = previous_mean["FIP"]-previous_mean["평균자책점"]
    except ZeroDivisionError:
        row['F-E'] = 0
        return row
    row['F-E'] = fe
    return row
```

```python
team_pitcher = team_pitcher.apply(fe,axis=1)
team_pitcher["F-E"]
```

```python
team_pitcher["F-E"].mean()
```

```python
pitcher = pitcher.apply(perFE,axis=1)
pitcher["F-E"]
```

```python
pitcher["F-E"].mean()
```

```python
def re(row):
    previous = team_pitcher[(team_pitcher['시즌'] == row['시즌']) & (team_pitcher['팀코드'] == row['팀코드']) & (team_pitcher['일자'] < row['일자'])]
    if len(previous) == 0:
        row['R-E'] = 0
        return row
    previous_sum = previous.mean()
    try:
        fe = previous_sum["RA/9"]-previous_sum["평균자책점"]
    except ZeroDivisionError:
        row['R-E'] = 0
    row["R-E"] = fe
    return row
def perRE(row):
    previous = pitcher[pitcher['선수코드'] == row['선수코드']].copy()
    previous = previous[(previous['시즌'] == row['시즌']) &(previous['일자'] < row['일자'])]
    previous_sum = previous.mean()
    if len(previous) == 0:
        row['R-E'] = 0
        return row
    try:
        re = previous_sum["RA/9"]-previous_sum["평균자책점"]
    except ZeroDivisionError:
        row['R-E'] = 0
        return row
    row['R-E'] = re
    return row
```

```python
team_pitcher = team_pitcher.apply(re,axis=1)
team_pitcher["R-E"]
```

```python
pitcher = pitcher.apply(perRE,axis=1)
pitcher["R-E"]
```

```python
team_pitcher["R-E"].mean()
```
