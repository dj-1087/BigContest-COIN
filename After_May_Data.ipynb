```
import pandas as pd
PATH = "/content/gdrive/My Drive/Colab Notebooks/2차_최종_데이터/"
pitcher_r_vs = pd.read_csv(PATH+"2차_최종_최근_상대_팀투수.csv")
hitter_r_vs = pd.read_csv(PATH+"2차_최종_최근_상대_팀타자.csv")
pitcher_r = pd.read_csv(PATH+"2차_최종_최근_팀투수.csv")
hitter_r = pd.read_csv(PATH+"2차_최종_최근_팀타자.csv")    
```

```
def after_may(df):
    team_list = []
    for year in range(2016, 2021):
        a = df[(df['일자'] > '{}-05-10'.format(year)) & (df['시즌'] == year)]
        team_list.append(a)
    df = pd.concat(team_list)
    return df
```

```
pitcher_r_vs = after_may(pitcher_r_vs)
pitcher_r_vs = pitcher_r_vs.reset_index()
pitcher_r = after_may(pitcher_r)
pitcher_r = pitcher_r.reset_index()
hitter_r_vs = after_may(hitter_r_vs)
hitter_r_vs = hitter_r_vs.reset_index()
hitter_r = after_may(hitter_r)
hitter_r = hitter_r.reset_index()
```
