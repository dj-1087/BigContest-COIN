### 전체 표 훑어보기

```python
import pandas as pd
PATH_IP = "C:/Users/home/DA(inflearn)/individual_pitcher/2020빅콘테스트 문제데이터(데이터분석분야-퓨처스리그)/2020빅콘테스트 데이터분석분야-퓨쳐스리그_스포츠투아이_제공데이터(.CSV)_시즌별, 시트별 구분/2020빅콘테스트_스포츠투아이_제공데이터_개인투수_"
PATH_P = "C:/Users/home/DA(inflearn)/individual_pitcher/2020빅콘테스트 문제데이터(데이터분석분야-퓨처스리그)/2020빅콘테스트 데이터분석분야-퓨쳐스리그_스포츠투아이_제공데이터(.CSV)_시즌별, 시트별 구분/2020빅콘테스트_스포츠투아이_제공데이터_선수_"

doc_ip = None
doc_p = None
for i in [2016, 2017, 2018, 2019, 2020]:
    doc_i = pd.read_csv(PATH_IP+str(i)+".csv", encoding='utf-8-sig')
    doc_ip = pd.concat([doc_ip, doc_i], axis=0)
    doc_j = pd.read_csv(PATH_P+str(i)+".csv", encoding='CP949')
    doc_p = pd.concat([doc_p, doc_j], axis=0)
doc_p.columns = ["GYEAR","P_ID","NAME","T_ID","POSITION","AGE_VA","MONEY"]
doc = pd.merge(doc_ip, doc_p, on="P_ID", how="left")
doc
```
