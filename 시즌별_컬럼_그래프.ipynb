```
def convert_days(row):
    row = row.split("-")
    row = int(str(row[1])+str(row[2]))
    return row
team_hitter["일자"] = team_hitter["일자"].apply(convert_days)
team_hitter["일자"]
```

```
team_hitter_mean = team_hitter.groupby(by=["시즌","일자"]).mean()
team_hitter_mean = team_hitter_mean.reset_index()
team_hitter_mean
```

```
team_hitter_
```

```
seasons = [2016,2017,2018,2019,2020]
for i in range(5):
    seasons[i] = team_hitter_mean[team_hitter_mean["시즌"]==seasons[i]]
seasons[1]
```

```
import plotly.graph_objects as go
fig = go.Figure()
for i in range(len(seasons)):
    fig.add_trace(go.Scatter(x=seasons[i].iloc[:,1], y=seasons[i].iloc[:,10],
                             mode='lines+markers',
                             name=str(seasons[i].iloc[0,0])))
fig.update_layout(
    {
        "title": {
            "text": "<b>시즌/일자</b> 별 "+str(seasons[i].columns[10])+" 그래프 <b><mean></b>",
            "x": 0.5,
            "y": 0.9,
            "font": {
                "size": 20
            }
        },
        "showlegend": True,
        "xaxis": {
            "title": "일자",
            "showticklabels":True,
            "dtick":1 
            
        },
        "yaxis": {
            "title": "수치"
        },

    }
)

fig.show()
```

```
def convert_months(row):
    row = int(row/100)
    return row
team_hitter["일자"] = team_hitter["일자"].apply(convert_months)
team_hitter["일자"]
```

```
team_hitter_mean = team_hitter.groupby(by=["시즌","일자"]).mean()
team_hitter_mean = team_hitter_mean.reset_index()
team_hitter_mean
```

```
seasons = [2016,2017,2018,2019,2020]
for i in range(5):
    seasons[i] = team_hitter_mean[team_hitter_mean["시즌"]==seasons[i]]
seasons[1]
```

```
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
colors=["red","yellow","green","blue","purple"]
fig = make_subplots(rows=len(team_hitter_mean.columns), cols=1)
for j in range(3,len(team_hitter_mean.columns)):
    for i in range(len(seasons)):
        np.random.seed(2)
        fig.add_trace(go.Scatter(x=seasons[i].iloc[:,1], y=seasons[i].iloc[:,j],
                                 mode='lines+markers', line=dict(color=colors[i]),
                                 name=str(seasons[i].iloc[0,0])+),row=j+1,col=1)
fig.update_layout(
    {
        "title": {
            "text": "<b>시즌/일자</b> 별 컬럼 그래프 <b><mean></b>",
            "font": {
                "size": 20
            }
        },
        "showlegend": True,
        "xaxis": {
            "title": "일자",
            "showticklabels":True,
            "dtick":1 
            
        },
        "yaxis": {
            "title": "수치"
        },
        "width": 1000,
        "height": 10000,
    }
)

fig.show()
```
