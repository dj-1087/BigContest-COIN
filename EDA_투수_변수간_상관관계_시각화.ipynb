```
import pandas as pd
PATH = "/content/gdrive/My Drive/Colab Notebooks/1차_최종_데이터(08_29)/"
team_pitcher = pd.read_csv(PATH+"1차_최종_팀투수.csv", encoding="CP949")
per_pitcher = pd.read_csv(PATH+"1차_최종_개인투수.csv")
team_pitcher_r = pd.read_csv(PATH+"1차_최종_최근_팀투수.csv")
team_pitcher_vs = pd.read_csv(PATH+"1차_최종_상대_팀투수.csv")
final_common = pd.read_csv(PATH+"1차_최종_공통.csv")
```

```
final_common = final_common[["G_ID","T_ID","VS_T_ID","aR"]]
team_pitcher = pd.merge(team_pitcher,final_common, on=["G_ID","T_ID","VS_T_ID"], how="left")
team_pitcher
```

```
import plotly.graph_objects as go
import plotly.express as px

fig = px.scatter(    
        x=team_pitcher['aR'], y=team_pitcher['whip']
    )


fig.update_layout(
    {
        "title": {
            "text": "Graph with go.Scatter",
            "font": {
                "size": 15
            }
        },
        "showlegend": True,
        "xaxis": {
            "title": "평균 실점"
        },
        "yaxis": {
            "title": "whip"
        }
    }
)
```

```
team_list = []
for year in range(2016, 2021):
  a = team_pitcher[(team_pitcher['GDAY_DS'] > '{}-05-10'.format(year)) & (team_pitcher['GYEAR'] == year)]
  team_list.append(a)
team_pitcher = pd.concat(team_list)
```

```
import plotly.graph_objects as go
import plotly.express as px

fig = px.scatter(    
        x=team_pitcher['aR'], y=team_pitcher['whip'],trendline = "ols",trendline_color_override="#ff6347"
    )


fig.update_layout(
    {
        "title": {
            "text": "Graph with go.Scatter",
            "font": {
                "size": 15
            }
        },
        "showlegend": True,
        "xaxis": {
            "title": "평균 실점"
        },
        "yaxis": {
            "title": "whip"
        }
    }
)

fig.show()
```
